```python
data = np.random.randn(100,2)@np.array([[1,0.5],[0.5,1]])+np.array([1,4])

plt.figure()
plt.scatter(data[:,0], data[:,1], marker=".")
plt.grid()
plt.axis("equal")
plt.show()
```

```python
g = [LinearFunction(1/2, i) for i in [2,-1,4,6]]
plt.figure()
plt.scatter(data[:,0], data[:,1], marker=".")
for i in range(len(g)):
    plt.plot(data[:,0], g[i](data[:,0]))
plt.grid()
plt.legend([str(f) for f in g])
plt.show()
```
