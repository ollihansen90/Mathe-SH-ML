```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    output = np.abs(x)
    d_output = np.sign(x)
    return output, d_output

base = np.linspace(-4, 4)
output, d_output = f(base)

plt.figure()
plt.plot(base, output)
plt.plot(base, d_output)
plt.show()
```

```python
start = 1
n_schritte = 10
lr = 0.6
schritte = np.zeros(n_schritte)
schritte[0] = start

for i in range(1, n_schritte):
    # x = x+alpha*(-df(x))
    wert, ableitung = f(schritte[i-1])
    schritte[i] = schritte[i-1] - lr * ableitung

plt.figure()
plt.plot(base, output)
plt.plot(base, d_output)
plt.plot(schritte, f(schritte)[0], "o-")
plt.show()
print(schritte)
# SGD: Stochastic Gradient Descent
```
